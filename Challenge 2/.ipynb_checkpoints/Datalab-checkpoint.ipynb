```python
import pandas, matplotlib.pyplot as plt
```

```python
df = pandas.read_excel('ExportDATALAB_feb2018.xlsx', index_col='Datum')
df_NO3 = df.loc[pandas.notnull(df['NO3[mg/L]']), ['Naam', 'NO3[mg/L]']]
df_NO3.loc[df_NO3['NO3[mg/L]'].astype(str).str.contains('<'), 'NO3[mg/L]']=0
df_NO3['NO3[mg/L]']=pandas.to_numeric(df_NO3['NO3[mg/L]'])
```

```python
fig, ax = plt.subplots(figsize=[16,9])
df_NO3['NO3[mg/L]'].plot(style='r.', ax=ax)
#df_NO3.groupby(df_NO3['NO3[mg/L]'].index.year).mean().plot(style='b-', ax=ax)
```

```python
df_NO3.loc[df_NO3['NO3[mg/L]'].astype(str).str.contains('<'), 'NO3[mg/L]']=0

df_CA = df.loc[pandas.notnull(df['Ca[mg/L]']), ['Naam', 'Ca[mg/L]']]
df_CA.loc[df_CA['Ca[mg/L]'].astype(str).str.contains('<'), 'Ca[mg/L]']=0
df_CA['Ca[mg/L]']=pandas.to_numeric(df_CA['Ca[mg/L]'])
```

```python
df_CA.plot(style='r.')
```

```python
fig, ax = plt.subplots(figsize=[16, 9])
df_CA.groupby(df_CA.index.year).mean().plot(ax=ax)
plt.ylim(0)
df_NO3.groupby(df_NO3.index.year).mean().plot(ax=ax, secondary_y=True)
plt.xlim(1971, 2017)
plt.ylim(0)
```
